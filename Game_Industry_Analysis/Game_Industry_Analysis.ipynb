```python
def percent_hbar(df, old_threshold=None):
    percent_of_nulls = (df.isnull().sum()/len(df)*100).sort_values().round(2)
    threshold = percent_of_nulls.mean()
    ax = percent_of_nulls.plot(kind='barh', figsize=(17, 8), title='% of NaN (из {} строк)'.format(len(df)), 
                               color='#86bf91', legend=False, fontsize=12)
    ax.set_xlabel('Count of NaN')
    dict_percent = dict(percent_of_nulls)
    i = 0
    for k in dict_percent:
        color = 'blue'
        if dict_percent[k] > 0:
            if dict_percent[k] > threshold:
                color = 'red'
            ax.text(dict_percent[k]+0.1, i + 0.09, str(dict_percent[k])+'%', color=color, 
                    fontweight='bold', fontsize='large')
        i += 0.98
    if old_threshold is not None:
        plt.axvline(x=old_threshold,linewidth=1, color='r', linestyle='--')
        ax.text(old_threshold+0.3, 10, '{0:.2%}'.format(old_threshold/100), color='r', fontweight='bold', fontsize='large')
        plt.axvline(x=threshold,linewidth=1, color='green', linestyle='--')
        ax.text(threshold+0.3, 7, '{0:.2%}'.format(threshold/100), color='green', fontweight='bold', fontsize='large')
    else:
        plt.axvline(x=threshold,linewidth=1, color='r', linestyle='--')
        ax.text(threshold+0.3, 7, '{0:.2%}'.format(threshold/100), color='r', fontweight='bold', fontsize='large')
    ax.set_xlabel('')
    return ax, threshold

plot, threshold = percent_hbar(df)
```

* В столбцах Genre, Year_of_Release, Name имеются пропущенные значения, но значительно меньше.

Уберем из датафрейма данные с пропущенными значениями, чтобы shap мог четче выдавать результаты

```python
df_cat_boost = df[~(df["Year_of_Release"].isna()) & ~(df["Critic_Score"].isna()) & ~(df["User_Score"].isna())]
```

Создадим обучающую выборку, валидационную и тестовую выборку. 

```python
import shap
shap.initjs()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test  =train_test_split(df_cat_boost,train_size = 0.6, random_state=42)
```

```python
test, val  =train_test_split(df_cat_boost,train_size = 0.5, random_state=42)
```

Зададим фичи и предикты, однако мы не будем использовать данные от продаж, так как они будут больше всего влиять на продажи по всему миру т.к. данная переменная складывается из них

```python
X = ["Platform","Year_of_Release", "Genre","Critic_Score", "User_Score","Rating"]

cat_features = ["Platform","Genre","Rating"]

y = ['World_sales']  
```

Сразу объединим обучающую и тестовую выборку для более лучшей модели

```python
train_full = pd.concat([train, val])
```

```python
from catboost import Pool

train_full_D = Pool(data=train_full[X],
                  label=train_full[y],
                  cat_features=cat_features
                 )

valid_data = Pool(data=val[X],
                  label=val[y],
                  cat_features=cat_features
                 )
```

```python
params = {'verbose' : 500,
          'eval_metric' : 'MAPE',
          'random_seed':42,
          'loss_function': 'MAE',
          "iterations" : 5000,
          'learning_rate': 0.07}
```

```python
model = CatBoostRegressor(**params)
```

```python
model.fit(train_full_D, eval_set=valid_data)
```

Сделаем предсказание на тестовой выборке, чтобы оценить модель

```python
test['World_sales_pred'] = model.predict(test[X])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["World_sales"],test["World_sales_pred"])
```

Как мы видим, наша модель в среднем ошибается на значение 0.46

Сделаем предикт на всю выборку

```python
df_cat_boost['World_sales_pred'] = model.predict(df_cat_boost[X])
```

Успользуем для анализа данные для которых разница между предиктом (World_sales_pred) и настоящими доходами по миру (World_sales) меньше 0.05

```python
df_true =  df_cat_boost[abs(df_cat_boost["World_sales_pred"] - df_cat_boost["World_sales"]) < 0.05]
```

```python
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(df_true[X])
```

Для даленьйшего анализа с использованием shap необходимо сказать пару слов о самом инструменте. Данный инструмент анализирует предсказанные значения мировых продаж (World_sales_pred) моделью CatBoost и считывает предикты, которые повлияли в положительную или в отрицательную сторону на определенное числовое значение

```python
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[0,:], df_true[X].iloc[0,:])
```

Важно не получить оценку ниже 4-5

Далее с 51 до 65 появляется плато, когда доходы падают или растут не сильно. После 65 наблюдается очень быстрый рост, здесь можно отметить, что с последующим прибавлением по 5 пунктам к Critic_Score ускаряет темпы роста доходов. На пункте 85-90 рост немного замедляется и на последнем отрезке имеет максимально-рекордное значение. Для Северо-Американского рынка важно получить для игры оценку 65-70 для значительного роста дохода и не получить оценку рейтинга ниже 50-55. Чем оценка больше, тем лучше для продаж

После 7.5 также имеется рост на некоторых участках + мизерная стагнация на участке 8 - 8.5. Для рынка Северной Америки для игры необходимо добиться пользовательских оценок не ниже 4.5 последующие значения User_Score не сильно влияют на продажи, однако после 7.5 это сильно поможет увеличить прибыль
